# 6. Manipulando os dados

### 6.1 Criando colunas numéricas

```python
# Atribuindo o endereço da base de dados para uma variável
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'

# Lendo com separador de colunas
dados_dev = pd.read_csv(url, sep = ';')
```

```python
# Lendo o dataframe
dados_dev.head()
```

```python
# Tratando os valores nulos
dados_dev.fillna(0, inplace = True)
dados_dev
```

```python
# Somando as colunas de valor e condominio, e criando a coluna de valor por mês
dados_dev['Valor_por_mes'] = dados_dev['Valor'] + dados_dev['Condominio']
dados_dev.head()
```

```python
# Criando a coluna de valor por ano
dados_dev['Valor_por_ano'] = dados_dev['Valor_por_mes'] * 12 + dados_dev['IPTU']
dados_dev.head()
```

### 6.2 Criando colunas categóricas

```python
dados_dev['Descricao'] = dados_dev['Tipo'] + ' em ' + dados_dev['Bairro']
dados_dev.head()
```

```python
# Criando a coluna com o resto das informações
dados_dev['Descricao'] = dados_dev['Tipo'] + ' em ' + dados_dev['Bairro'] + ' com ' + \
                         dados_dev['Quartos'].astype(str) + ' quarto(s) ' + \
                         ' e ' + dados_dev['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
# Conferindo a adição da coluna
dados_dev.head()
```

```python
# Adicionando coluna determinando se possui suite ou não
dados_dev['Possui_suite'] = dados_dev['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados_dev.head()
```

```python
# Salvando arquivo para o time de dev
dados_dev.to_csv('dados_completos_dev.csv', index = False, sep = ';')
```
